```python
import pandas as pd
import ast

def extract_transactions(csv_file, required_role):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Filter the dataframe for normal transactions of the required role
    filtered_df = df[(df['role'] == required_role) & (df['class'] == 0)]
    
    # Extract the query column and convert string representations of lists to actual lists
    transactions = filtered_df['query'].apply(ast.literal_eval).tolist()
    
    # Format the transactions
    formatted_transactions = []
    for transaction in transactions:
        # Join operations with '-1' and add double '-1' at the end
        formatted_transaction = ' -1 '.join(transaction) + ' -1 -1'
        formatted_transactions.append(formatted_transaction)
    
    # Join all transactions with newline
    output = '\n'.join(formatted_transactions)
    
    return output

def write_output(output, filename):
    with open(filename, 'w') as f:
        f.write(output)

# Example usage
csv_file = '/kaggle/input/temp-data/training_queries_for_22.5k.csv'
required_role = req_role  # Change this to the desired role
output_file = 'extracted_transactions.txt'

output = extract_transactions(csv_file, required_role)
write_output(output, output_file)
print(f"Extracted transactions have been written to {output_file}")
```

```python
import sys
import time
import re
from collections import defaultdict

# ======== INTEGRATED PDATA ========
class processingData:
    def read_file(self, readfilename, max_lines=None):
        lines_s = []
        with open(readfilename, 'r') as f:
            for i, line in enumerate(f):
                if max_lines and i >= max_lines:
                    break
                lines_s.append(line.strip())
        return lines_s

    def item_sotrd(self, items):
        seen = set()
        return sorted([x for x in items if not (x in seen or seen.add(x))]), []

    def item_to_dict(self, sort_items, len_lines, S):
        for item in sort_items:
            S[item] = [[] for _ in range(len_lines)]
        return S

    def split_array(self, lines):
        lines = lines.replace('  ', ' ')
        items_array = lines.strip().split(' ')
        s_array = [[]]
        i = 0
        for item in items_array:
            if item != '-1':
                s_array[i].append(item)
            else:
                i += 1
                s_array.append([])
        return items_array, s_array

    def Statistics_items(self, lines, items):
        items_array = lines.replace('  ', ' ').strip().split(' ')
        return list(set(items + [x for x in items_array if x != '-1']))

    def General_Sn(self, lines_s, S):
        for i, line in enumerate(lines_s):
            _, s_array = self.split_array(line)
            for k, subset in enumerate(s_array):
                for item in set(subset):
                    if item in S:
                        S[item][i].append(k)
        return S

    def datap(self, filename, S, max_lines=20):
        lines_s = self.read_file(filename, max_lines)
        items = []
        for line in lines_s:
            items = self.Statistics_items(line, items)
        sort_items, _ = self.item_sotrd(items)
        self.item_to_dict(sort_items, len(lines_s), S)
        self.General_Sn(lines_s, S)
        return len(lines_s), S, sort_items
```

```python
#main rule miner

from itertools import combinations
import ast

def parse_frequent_patterns(file_path):
    frequent_patterns = []
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith("Number of"):
                continue
            if " - Support: " in line:
                pattern_str, support_str = line.split(" - Support: ")
                pattern = ast.literal_eval(pattern_str.strip())
                support = int(support_str.strip())
                # Ensure single-item patterns are wrapped in lists for consistency
                if isinstance(pattern, list) and all(isinstance(x, str) for x in pattern):
                    pattern = [[x] for x in pattern]
                frequent_patterns.append((pattern, support))
    return frequent_patterns

file_path = r"/kaggle/working/frequent_patterns_3.txt"
frequent_patterns = parse_frequent_patterns(file_path)

total_transactions = 5020
pattern_support = {}
for pattern, support in frequent_patterns:
    pattern_support[tuple(map(tuple, pattern))] = support / total_transactions

min_conf = 0.3
min_len = 2

def generate_rules(pattern):
    rules = []
    pattern = list(map(list, pattern))
    n = len(pattern)
    for i in range(1, n):
        for antecedent_indices in combinations(range(n), i):
            antecedent = [pattern[idx] for idx in antecedent_indices]
            consequent = [pattern[idx] for idx in range(n) if idx not in antecedent_indices]
            if antecedent and consequent:
                rules.append((antecedent, consequent))
    return rules

rules_set = []
for pattern, _ in frequent_patterns:
    pattern = list(map(list, pattern))
    pattern_tuple = tuple(map(tuple, pattern))
    sup = pattern_support[pattern_tuple]
    
    if len(pattern) < min_len:
        continue
    
    possible_rules = generate_rules(pattern)
    
    for antecedent, consequent in possible_rules:
        antecedent_tuple = tuple(map(tuple, antecedent))
        if antecedent_tuple in pattern_support:
            sup_antecedent = pattern_support[antecedent_tuple]
            if sup_antecedent > 0:
                confidence = sup / sup_antecedent
                if confidence >= min_conf:
                    rules_set.append({
                        'antecedent': antecedent,
                        'consequent': consequent,
                        'support': sup,
                        'confidence': confidence
                    })

# Write the rules to a text file
output_file_path = "dependency_rules_new.txt"
with open(output_file_path, 'w') as file:
    file.write("Generated Dependency Rules (No min_sup threshold):\n")
    file.write("Association Rule\t\t\tSupport\tConfidence\n")
    for rule in rules_set:
        antecedent_str = "{" + ", ".join([",".join(item) for item in rule['antecedent']]) + "}"
        consequent_str = "{" + ", ".join([",".join(item) for item in rule['consequent']]) + "}"
        file.write(f"{antecedent_str} -> {consequent_str}\t{rule['support']:.6f}\t{rule['confidence']:.6f}\n")
    file.write(f"\nTotal number of rules generated: {len(rules_set)}\n")

print(f"Rules have been written to {output_file_path}")
```

```python
def parse_rule(rule_string):
    # Split the rule into antecedent and consequent parts
    parts = rule_string.split('->')
    if len(parts) != 2:
        return None
    
    # Parse antecedent (left side)
    antecedent_str = parts[0].strip().replace('{', '').replace('}', '')
    antecedent_items = set(item.strip() for item in antecedent_str.split(',') if item.strip())
    
    # Parse consequent (right side)
    consequent_str = parts[1].strip().replace('{', '').replace('}', '')
    consequent_items = set(item.strip() for item in consequent_str.split(',') if item.strip())
    
    return [antecedent_items, consequent_items]

def convert_rules_to_array(file_path):
    rules_array = []
    
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()
            
            # Skip header lines and process only rule lines
            start_processing = False
            for line in lines:
                line = line.strip()
                
                # Start processing after the header
                if line.startswith('{') and '->' in line:
                    start_processing = True
                    
                if start_processing and '->' in line:
                    # Extract the rule part before the support/confidence values
                    rule_part = line.split('\t')[0]
                    parsed_rule = parse_rule(rule_part)
                    if parsed_rule:
                        rules_array.append(parsed_rule)
                        
                # Stop processing when we reach the total count
                if line.startswith('Total number of rules'):
                    break
                    
        return rules_array
    
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found")
        return []
    except Exception as e:
        print(f"Error reading file: {str(e)}")
        return []


rules = convert_rules_to_array('dependency_rules_new.txt')
print(rules[:20])

# def combine_rule_elements_ordered(rules):
#     """
#     Combines the antecedent and consequent of each rule into a single list,
#     preserving the sequence: antecedent elements followed by consequent elements.
    
#     Args:
#         rules: List of rules where each rule is [antecedent_set, consequent_set]
    
#     Returns:
#         List of tuples: (rule_id, combined_list) where combined_list preserves order
#     """
#     combined_rules = []
#     for i, rule in enumerate(rules):
#         # Convert sets to lists if they aren't already
#         antecedent = list(rule[0])
#         consequent = list(rule[1])
        
#         # Combine while preserving order: antecedent followed by consequent
#         combined_list = antecedent + consequent
        
#         # Store as tuple with rule ID for reference
#         combined_rules.append((combined_list))
    
#     return combined_rules

# combine antecedent and consequent 
# rules_comb = combine_rule_elements_ordered(rules) 

# print(rules_comb)
```

```python
import json
import pandas as pd
import ast
from tqdm import tqdm  # Import tqdm for progress tracking

def lcs_length(seq1, seq2):
    m, n = len(seq1), len(seq2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if seq1[i-1] == seq2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[m][n]

def stage_0_similarity(rule, sequence):
    lcs_len = lcs_length(rule, sequence)
    total_len = len(rule) + len(sequence)
    return (2 * lcs_len) / total_len if total_len > 0 else 0.0

def stage_1_similarity(rule, sequence, alpha=1.0, beta=1.0):
    rule_set = set(rule)
    seq_set = set(sequence)
    intersection = len(rule_set.intersection(seq_set))
    rule_minus_seq = len(rule_set - seq_set)
    seq_minus_rule = len(seq_set - rule_set)
    denominator = intersection + (alpha * rule_minus_seq) + (beta * seq_minus_rule)
    return intersection / denominator if denominator > 0 else 0.0

def stage_3_similarity(rule, sequence):
    max_len = max(len(rule), len(sequence))
    rule_padded = rule + [None] * (max_len - len(rule))
    seq_padded = sequence + [None] * (max_len - len(sequence))
    
    hamming_distance = sum(1 for r, s in zip(rule_padded, seq_padded) if r != s)
    
    return 1 - (hamming_distance / max_len) if max_len > 0 else 0.0

def parse_rules(file_path):
    rules = []
    try:
        with open(file_path, 'r') as f:
            lines = f.readlines()
        for line in lines[1:]:
            parts = line.strip().split('\t')
            if len(parts) >= 3:
                rule_str = parts[0]
                if '->' in rule_str:
                    items = rule_str.split('->')
                    if len(items) == 2:
                        antecedent = [item.strip().strip('{}"') for item in items[0].split(',') if item.strip()]
                        consequent = [item.strip().strip('{}"') for item in items[1].split(',') if item.strip()]
                        full_rule = antecedent + consequent
                        try:
                            support = float(parts[1])
                            confidence = float(parts[2])
                            rules.append((full_rule, support, confidence))
                        except ValueError:
                            pass
    except Exception as e:
        print(f"Error parsing rules file: {e}")
    return rules

def load_transactions(file_path):
    transactions = []
    classes = []
    predicted_classes = []
    try:
        df = pd.read_csv(file_path)
        if 'query' not in df.columns:
            print("Error: 'query' column not found in CSV")
            return transactions, classes, predicted_classes
        
        df['query'] = df['query'].apply(ast.literal_eval)  
        transactions = df['query'].tolist()
        
        # Extract class and predicted_class
        if 'class' in df.columns:
            classes = df['class'].tolist()
        if 'predicted_class' in df.columns:
            predicted_classes = df['predicted_class'].tolist()
            
    except Exception as e:
        print(f"Error loading transactions: {e}")
    return transactions, classes, predicted_classes

def calculate_similarities(rule_file, transaction_file, w0=1.0, w1=1.0, w3=1.0):
    rules = parse_rules(rule_file)
    transactions, classes, predicted_classes = load_transactions(transaction_file)
    
    if not rules or not transactions:
        return []

    best_results = {}
    
    # Use tqdm to track progress of transactions
    for trans_idx, transaction in tqdm(enumerate(transactions), total=len(transactions), desc="Processing Transactions"):
        for rule_items, support, confidence in rules:
            stage_0 = stage_0_similarity(rule_items, transaction)
            stage_1 = stage_1_similarity(rule_items, transaction)
            stage_3 = stage_3_similarity(rule_items, transaction)
            
            total_weight = w0 + w1 + w3
            srs = (w0 * stage_0 + w1 * stage_1 + w3 * stage_3) / total_weight if total_weight != 0 else 0.0
            
            if trans_idx not in best_results or srs > best_results[trans_idx]['srs']:
                best_results[trans_idx] = {
                    'transaction': transaction,
                    'stage_0_similarity': stage_0,
                    'stage_1_similarity': stage_1,
                    'stage_3_similarity': stage_3,
                    'srs': srs,
                    'best_rule': rule_items,
                    'support': support,
                    'confidence': confidence,
                    'class': classes[trans_idx] if trans_idx < len(classes) else None,
                    'predicted_class': predicted_classes[trans_idx] if trans_idx < len(predicted_classes) else None
                }
                
    return list(best_results.values())

if __name__ == "__main__":
    rule_file = "/kaggle/working/dependency_rules_new.txt"
    transaction_file = "/kaggle/working/transactions_role_0.csv"
    
    results = calculate_similarities(rule_file, transaction_file, w0=1.0, w1=1.0, w3=1.0)
    
    results_df = pd.DataFrame(results)[['transaction', 'stage_0_similarity', 'stage_1_similarity', 
                                     'stage_3_similarity', 'srs', 'class', 'predicted_class']]
    
    results_df.to_csv('max_srs_per_transaction.csv', index=False)
    print("Results saved to 'max_srs_per_transaction.csv'")
```
